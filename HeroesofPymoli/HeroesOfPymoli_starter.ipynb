```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
file_to_load = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# Total # of Players
player_count = purchase_data_df["SN"].count()
player_count

total_players = pd.DataFrame({"Total Number of Players": [player_count]})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Unique Items
unique_items = len(purchase_data_df["Item ID"].unique())
unique_items

# Average Purchase Price
average_price = purchase_data_df["Price"].mean()
average_price

# Total Number of Purchases
total_purchases = purchase_data_df["Purchase ID"].count()
total_purchases

# Total Revenue
total_revenue = purchase_data_df["Price"].sum()
total_revenue

# Display
purchase_analysis_df = pd.DataFrame([{"Number of Unique Items": unique_items,
                                      "Average Purchase Price": average_price,
                                      "Total Number of Purchases": total_purchases,
                                      "Total Revenue": total_revenue}])

purchase_analysis_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demo_df = pd.DataFrame(purchase_data_df["Gender"].value_counts())
gender_demo_df

percent_of_players = (purchase_data_df["Gender"].value_counts()/player_count)*100
percent_of_players

gender_demo_df["Percentage of Players"] = percent_of_players
gender_demo_df.rename(columns={"Gender":"Total Count"})


```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchased_data_df = purchase_data_df.groupby(["Gender"])
gender_purchased_data_df.head()

# Purchase Count
gender_purchased_data_df["Purchase ID"].count().head()

# Average Purchase Price
avg_purchase_price = gender_purchased_data_df["Price"].mean()
avg_purchase_price

# Total Purchase Value
purchase_value = gender_purchased_data_df["Price"].sum()
purchase_value

# Average Purchase Total
avg_purchase_total = purchase_value/gender_purchased_data_df["Purchase ID"].count()
avg_purchase_total

org_gender_purchased_data_df = pd.DataFrame(gender_purchased_data_df["Purchase ID"].count())
org_gender_purchased_data_df["Average Purchase Price"] = avg_purchase_price 
org_gender_purchased_data_df["Total Purchase Value"] = purchase_value 
org_gender_purchased_data_df["Average Purchase Total"] = avg_purchase_total
org_gender_purchased_data_df

org_gender_purchased_data_df.rename(columns={"Purchase ID":"Purchase Count"})
```
